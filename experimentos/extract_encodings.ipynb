```python
import face_recognition
import os
from queue import PriorityQueue
import time
import pandas as pd
import json
import numpy as np
```

```python
encodings = []
for dir_name in os.listdir("../input/face-images/dataset/"):
    for file_name in os.listdir("../input/face-images/dataset/" + dir_name):
        image = face_recognition.load_image_file("../input/face-images/dataset/" + dir_name + "/" + file_name)
        faces = face_recognition.face_encodings(image)
        #print("It has " + str(len(faces)) + " faces in " + file_name)
        if len(faces) > 0:
            encodings.append(np.append(faces[0],file_name))
```

```python
col_names = []
for i in range(len(encodings[0])-1):
    print(col_names.append(f"f_{i}"))
col_names.append("Path")
```

```python
sizes = []
for i in encodings:
    sizes.append(len(i))
print(f"Dims: {set(sizes)}")
```

```python
df = pd.DataFrame(data=encodings, columns=col_names)
```

```python
df.to_csv("/kaggle/working/encodings.csv", index=False, encoding="utf-8")
```
